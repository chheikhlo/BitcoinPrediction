### 1. Afficher une image
Afficher une image quelconque depuis le train en indiquant comme titre de l'image son
label (càd le target associé) ?

```python
def display_digit(index, data, labels):
    label = labels[index]
    image = data.iloc[index].values.reshape(28, 28)
    plt.imshow(image, cmap='gray')
    plt.title(f'Label: {label}')
    plt.axis('off')
    plt.show()

random_index = np.random.randint(0, len(train_df))
display_digit(random_index, train_df.drop('label', axis=1), train_df['label'])
```

### 2. Afficher chaque chiffre

Dans une même figure affiché une occrence au hasard de chaque chiffre

```python
fig, axes = plt.subplots(1, 10, figsize=(20, 2))

index_lignes = []

groupes = train_df.groupby('label')

# Itération sur chaque groupe
for chiffre, groupe in groupes:
    # Sélection aléatoire d'une ligne dans le groupe
    ligne_aleatoire = groupe.sample(n=1)
    print(f"{chiffre} est représenté {len(groupe)} fois")
    
    # Récupération de l'index de la ligne aléatoire et ajout à la liste
    index_ligne = ligne_aleatoire.index[0]
    index_lignes.append(index_ligne)
    
for i in range(len(index_lignes)):
    image = train_df.loc[index_lignes[i]].values[1:].reshape(28, 28)  # Skip the label
    
    axes[i].imshow(image, cmap='gray')
    axes[i].set_title(f'Label: {i}')
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

### 3. Afficher plusieurs fois le même chiffre
Pour visualiser les différentes façons d'écrire un 7, afficher les 9 1ères
images qui correspondent au chiffre 7 en les plaçant dans une même figure ?

```python
fig, axes = plt.subplots(1, 9, figsize=(18, 2))
sevens = train_df[train_df['label'] == 7].head(9)  

for i, (index, row) in enumerate(sevens.iterrows()):
    image = row.values[1:].reshape(28, 28)  
    axes[i].imshow(image, cmap='gray')
    axes[i].set_title(f'Index: {index}')
    axes[i].axis('off')
    
plt.tight_layout()
plt.show()
```

### 4. Représentant moyen

Afficher le représentant "moyen" de chaque chiffre ?

```python
fig, axes = plt.subplots(1, 10, figsize=(20, 2))

for digit in range(10):
    mean_image = train_df[train_df['label'] == digit].drop('label', axis=1).mean().values.reshape(28, 28)
    axes[digit].imshow(mean_image, cmap='gray')
    axes[digit].set_title(f'Average {digit}')
    axes[digit].axis('off')

plt.tight_layout()
plt.show()
```

On peut constater que les images sont tout de même très ressemblantes les unes des autres, cela pourra causer un problème de surapprentissage
